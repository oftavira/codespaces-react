```python
# First we read the data.sp file located one directory up
# inside the '/files' directory

import numpy as np

with open('../files/data.sp', 'r') as f:
    data = f.readlines()
```

```python
x = []
y = []

for line in data:
    line = line.replace('\n', '')
    x.append(float(line.split('\t')[0]))
    y.append(float(line.split('\t')[1]))
```

```python
x = np.array(x)
y = np.array(y)

# Now we plot the data

import matplotlib.pyplot as plt

plt.plot(x, y, 'o')
plt.show()
```

```python
x = x[::-1]
y = y[::-1]
```

```python
s = 50
xx = x[:s]
yy = y[:s]

# Now we plot the data with a line connecting the points

plt.plot(xx, yy, 'o')
plt.plot(xx, yy)
plt.show()
```
